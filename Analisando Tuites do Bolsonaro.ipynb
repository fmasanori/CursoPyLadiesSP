```python
# Vamos pegar agora um tuíte escrito por Jair Bolsonaro
frase = 'Há décadas a esquerda se infiltrou em nossas instituições e passou a promover sua ideologia travestida de posicionamentos técnicos. O decreto que assinei hoje extingue vagas para órgãos aparelhados no Conselho Nacional sobre Drogas e acaba com o viés ideológico nas discussões.'
# Se quisermos colocar tudo em minúsculo:
print (frase.lower())
# Se quisermos colocar tudo em maiúsculo:
print (frase.upper())
# Se quisermos contar quantas vezes aparece uma determinada palavra:
print (frase.count('ideologia'))
# Se quisermos substituir uma palavra ou outra:
print (frase.replace('esquerda', 'direita'))
```

### Dicionário

```python
dados['Partido'] = 'PSL'
dados
```

```python
if dados['Partido'] == 'PSL':
    print ('O partido atual do Deputado Eduardo Bolsonaro é o', dados['Partido'])
```

```python
if dados['Partido'] != 'PT':
    print ('Eduardo Bolsonaro não é afiliado ao partido PT.')
else:
    print ('O partido atual do Deputado Eduardo Bolsonaro é o', dados['Partido'])
```

**Acessando Dados**
--------------------------
Existem várias formas de acessar dados.

Orgãos públicos costumam disponibilizar, em vários formatos, estes dados para download. Outras vezes é necessário fazer um programa para acessar estes dados. Existem várias formas de fazer isso. 

A mais fácil é utilizando uma API, Interface de Programação de Aplicações, um facilitador de pedidos de informação. 
Como um aplicativo de delivery de comida, onde pedimos de forma fácil o que queremos comer naquele momento.

```python
import requests

url = 'http://educacao.dadosabertosbr.com/api/escolas/buscaavancada?situacaoFuncionamento=1'
situação = '&energiaInexistente=on&aguaInexistente=on&esgotoInexistente=on'
resp = requests.get(url+situação).json()
print ('Número de Escolas em funcionamento sem energia, água e esgoto:', resp[0])
for x in resp[1]:
  print (x['nome'], x['cod'])
  print (x['cidade'], x['estado'], x['regiao'])
  print ()
```

Um portal, com dados antigos, fornece estas informações. 

Como conseguimos acessá-las? 

Usando a API descrita em [http://educacao.dadosabertosbr.com/api]. O endereço que acessamos é [http://educacao.dadosabertosbr.com/api/escolas/buscaavancada?situacaoFuncionamento=1&energiaInexistente=on&aguaInexistente=on&esgotoInexistente=on]. 

Talvez suas informações apareçam embaralhadas. Neste caso instale uma extensão no seu navegador, o *JSONView*, e tente novamente. Peça ajuda para instalar a extensão à monitora. 

Nosso acesso retorna um JSON, formato muitíssimo utilizado para retorno de APIs atualmente. Conseguimos traduzir o JSON para um dicionário. O resultado é uma Lista onde a posição zero dá o número de escolas. Já na posição 1 há as informações sobre elas. Este site é meio instável, e caso não consiga acessar vamos ao próximo passo. 

Algumas vezes não é trivial usar uma API, como no caso dos tuítes, por exemplo. Para acessar a API do Twitter é necessário registrar chaves de acesso, um processo demorado para este curso.
